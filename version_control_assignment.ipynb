changes to a project and share them on GitHub.

Answer:

1.clone the repositry - git clone < repository_url >

2.Create or switch to a new branch - git checkout -b branch_name

3.Make chnges in the code

4.add the changed files - git add

5.commit the changes - git commit -m "your commit message"

6.push the branch to github - git push origin branch_name

7.create a pull reques(pr) on github to merg chaanges into the main branch.
